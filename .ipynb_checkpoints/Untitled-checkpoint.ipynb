```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
m = 1
alpha = 1
R = 1
def a_x(x,y,m=m,alpha=alpha,R=R):
    return -alpha/m *((x/(x**2+y**2)**1.5)+(x-R)/((x-R)**2+y**2)**1.5) 
def a_y(x,y,m=m,alpha=alpha,R=R):
    return -alpha*y/m *((1/(x**2+y**2)**1.5)+1/((x-R)**2+y**2)**1.5)
```

```python
def rdot(r, t):
    x, y, v_x, v_y = r
    return np.array([v_x, v_y, a_x(x,y), a_y(x,y)])
```

```python
def trajektoria(v_x, v_y,n_t = int(10**4), maxt=10):
    r_0 = np.array([0.6,1,v_x,v_y])
    t = np.linspace(0,maxt,n_t)
    r = odeint(rdot, r_0, t)
    return r[:,0], r[:,1]
```

```python
def wykres(v_x, v_y,n_t = int(10**4),maxt = 10):
    x,y = trajektoria(v_x, v_y,n_t,maxt)
    plt.plot(x,y)
    plt.plot([0,1],[0,0],'ro')
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('y')
    title = r'$v_x(t=0) = ' + str(v_x) +  r' \frac{m}{s}, v_y(t=0) = ' + str(v_y) + r' \frac{m}{s}$'
    plt.title(title)
```

```python
wykres(0,0, 1e4,100)
```

```python
wykres(1,1,maxt=100)
```

```python
wykres(1,-1,maxt=100)
```

```python
wykres(-1,-1,maxt=100)
```

```python
wykres(1,0,maxt=100)
```

```python
wykres(0,1,maxt=100)
```

```python
wykres(0,-1,maxt=100)
```
